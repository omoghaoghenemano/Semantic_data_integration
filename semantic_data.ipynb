```python
def parse_xml(file):
    if isinstance(file, str):
        tree = etree.parse(file)
    elif isinstance(file, etree._ElementTree):
        tree = file
    else:
        raise TypeError("Invalid input type for parse_xml")
    return tree
```

```python
def xml_to_schema_tree(xml_doc):
    schema_tree = {}

    def build_tree(element, tree):
        children = list(element)
        if not children:
            tree[element.tag] = None
        else:
            tree[element.tag] = {}
            for child in children:
                build_tree(child, tree[element.tag])

    root = xml_doc.getroot()
    build_tree(root, schema_tree)
    return schema_tree

def get_node_names_from_schema_tree(tree, node_names=None):
    if node_names is None:
        node_names = set()

    for key, value in tree.items():
        node_names.add(key)
        if isinstance(value, dict):
            get_node_names_from_schema_tree(value, node_names)

    return node_names

# Extract node names from the schema tree

```

```python
def print_schema_tree(tree, indent=0):
    for key, value in tree.items():
        print(' ' * indent + key)
        if isinstance(value, dict):
            print_schema_tree(value, indent + 2)


```

```python
# Function to get paths from schema tree
def get_paths_from_schema_tree(tree, current_path="", paths=[]):
    for key, value in tree.items():
        # Append current key to the path
        current_path += "/" + key
        if isinstance(value, dict):
            # Recursively process nested dictionaries
            get_paths_from_schema_tree(value, current_path, paths)
        else:
            # Reached a leaf node, add current path to paths list
            paths.append(current_path)
        # Remove current key before returning to previous level
        current_path = current_path.rsplit("/", 1)[0]
    return paths

# Get paths from schema tree
movie_paths = get_paths_from_schema_tree(movie_tree)
mediated_paths = get_paths_from_schema_tree(mediated_tree)

# Print paths
print("movie_paths")
for path in movie_paths:
    print(path)

# Print paths
print("mediated_paths")
for path in mediated_paths:
    print(path)

```

```python
#TODO Add compute distance of interest
""" how to run
distance = SequenceBasedSimilarities()
distance.edit_distance('game', 'gamee')
"""
class SequenceBasedSimilarities():
    def __init__(self):
        pass

    def normalized_edit_distance(self,str1, str2):
      edit_distance = jellyfish.levenshtein_distance(str1, str2)
      max_length = max(len(str1), len(str2))

      if max_length == 0:  # To handle the case when both strings are empty
          return 0.0

      normalized_distance =  edit_distance / max_length
      return 1 - normalized_distance

    def jaro(self, str1, str2):
       return jellyfish.jaro_similarity(str1, str2)

    def jara_winkler(self, str1, str2):
       return jellyfish.jaro_winkler_similarity(str1, str2)


```

```python
#TODO compute similarities

class SimilaritiesMeasure():
    """ how to run
    similarity = SimilaritiesMeasure()
    similarity.{**method of interest""}('game', 'gamee')
    for example
    similarity.generalized_jaccard('game','gamee')
    """
    def __init__(self):
        pass

    def word2vec(self, str1, str2, glove_model_path):
        """ add model path"""
        model =  glove_model_path
        tokens1 = str1.split()
        tokens2 = str2.split()
        vec1 = sum(model[token] for token in tokens1 if token in model)
        vec2 = sum(model[token] for token in tokens2 if token in model)
      # Check if vec1 or vec2 is empty
        if isinstance(vec1, int) or isinstance(vec2, int):
            return 0  # Return 0 similarity if one of the vectors is empty

        # Compute cosine similarity if both vectors are non-empty
        return 1 - cosine(vec1, vec2)

    def setBasedSimilarities_jaccard(self, str1, str2):
        tokens1 = set(word_tokenize(str1.lower()))
        tokens2 = set(word_tokenize(str2.lower()))
        jaccard = len(tokens1 & tokens2) / len(tokens1 | tokens2)

        return jaccard

    def setBasedSimilarities_overlap(self, str1, str2):
        tokens1 = set(word_tokenize(str1.lower()))
        tokens2 = set(word_tokenize(str2.lower()))
        overlap = len(tokens1 & tokens2) / min(len(tokens1), len(tokens2))

        return overlap

    def Hybrid(self, str1, str2):
        # TF/IDF cosine similarity
        vectorizer = TfidfVectorizer().fit([str1, str2])
        tfidf_matrix = vectorizer.transform([str1, str2])
        tfidf_cosine_similarity = (tfidf_matrix * tfidf_matrix.T).A[0, 1]

        return tfidf_cosine_similarity

    def generalized_jaccard(self, str1, str2):
            vec1 = Counter(word_tokenize(str1.lower()))
            vec2 = Counter(word_tokenize(str2.lower()))
            intersection = sum(min(vec1[word], vec2[word]) for word in set(vec1) & set(vec2))
            union = sum(max(vec1[word], vec2[word]) for word in set(vec1) | set(vec2))
            return intersection / union

    def soft_tfidf(self, str1, str2):
            tfidf_vectorizer = TfidfVectorizer().fit([str1, str2])
            tfidf_matrix = tfidf_vectorizer.transform([str1, str2]).toarray()
            cosine_sim = np.dot(tfidf_matrix[0], tfidf_matrix[1]) / (np.linalg.norm(tfidf_matrix[0]) * np.linalg.norm(tfidf_matrix[1]))
            return cosine_sim
```

```python
# Function to compute Jaccard similarity
def jaccard_similarity(set1, set2):
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection / union if union != 0 else 0

# Function to match nodes based on their tags
def match_nodes(node1, node2):
    return 1 if node1 == node2 else 0

# Function to create a schema tree and extract paths
def create_schema_tree_and_extract_paths(xml_file):
    xml_doc = parse_xml(xml_file)
    schema_tree = xml_to_schema_tree(xml_doc)
    paths = get_paths_from_schema_tree(schema_tree)
    return paths

# Function to compute node similarity matrix
def node_similarity_matrix(schema1, schema2):
    matrix = defaultdict(dict)
    for path1, path2 in product(schema1, schema2):
        matrix[tuple(path1.split("/"))][tuple(path2.split("/"))] = jaccard_similarity(set(path1.split("/")), set(path2.split("/")))
    return matrix

# Function to compute path similarity
def path_similarity(schema1, schema2):
    similarities = []
    for path1, path2 in product(schema1, schema2):
        similarities.append(jaccard_similarity(set(path1.split("/")), set(path2.split("/"))))
    return similarities
```

```python
def calculate_similarities_hybrid_set_based_for_paths(columns1, columns2, type):
    """ types are word2vec, jaccard, soft_tfidf, hybrid, set_based_based_jaccard, set_based_based_overlap"""
    similarity_results = pd.DataFrame()
    similarity_measure = SimilaritiesMeasure()

    # Iterate over all pairs of columns and compute similarities
    for col1 in columns1:
        for col2 in columns2:
            if(type == "word2vec"):
                similarity = similarity_measure.word2vec(col1, col2, glove_model)
                similarity_results.loc[col1, col2] = similarity
            elif(type == "jaccard"):
                similarity = similarity_measure.generalized_jaccard(col1, col2)
                similarity_results.loc[col1, col2] = similarity
            elif(type == "soft_tfidf"):
                similarity = similarity_measure.soft_tfidf(col1, col2)
                similarity_results.loc[col1, col2] = similarity
            elif(type == "hybrid"):
                similarity = similarity_measure.Hybrid(col1, col2)
                similarity_results.loc[col1, col2] = similarity
            elif( type == "set_based_based_jaccard"):
                similarity = similarity_measure.setBasedSimilarities_jaccard(col1, col2)
                similarity_results.loc[col1, col2] = similarity
            elif(type == "set_based_based_overlap"):
                similarity = similarity_measure.setBasedSimilarities_overlap(col1, col2)
                similarity_results.loc[col1, col2] = similarity

    return similarity_results
```

```python
def calculate_similarities_sequence_based_for_paths(paths1, paths2, similarity_type):
    similarity_results = pd.DataFrame()
    similarity_measure = SequenceBasedSimilarities()

    # Iterate over all pairs of paths and compute similarities
    for path1 in paths1:
        for path2 in paths2:
            if similarity_type == "normalized_edit_distance":
                similarity = similarity_measure.normalized_edit_distance(path1, path2)
            elif similarity_type == "jaro":
                similarity = similarity_measure.jaro(path1, path2)
            elif similarity_type == "jara_winkler":
                similarity = similarity_measure.jara_winkler(path1, path2)
            else:
                similarity = 0

            similarity_results.loc[path1, path2] = similarity

    return similarity_results
```

```python
import pandas as pd

def create_ground_truth_df(original_nodes, mediated_nodes, ground_truth_node):
    """
    Create a ground truth DataFrame from original nodes, mediated nodes, and their mapping.

    Parameters:
    original_nodes (list): List of original node names.
    mediated_nodes (list): List of mediated node names.
    ground_truth_node (dict): Dictionary mapping mediated nodes to original nodes.

    Returns:
    pd.DataFrame: Ground truth DataFrame with mediated nodes as index and original nodes as columns.
    """
    # Ensure the input lists and dictionary are not empty
    if not original_nodes or not mediated_nodes or not ground_truth_node:
        raise ValueError("Input lists and dictionary must not be empty.")

    # Create an empty DataFrame
    ground_truth_df = pd.DataFrame(0, index=mediated_nodes, columns=original_nodes)

    # Populate the DataFrame using the updated ground truth mapping
    for mediated_node, original_node in ground_truth_node.items():
        if mediated_node in mediated_nodes and original_node in original_nodes:
            ground_truth_df.loc[mediated_node, original_node] = 1

    return ground_truth_df

# Example usage:
# original_nodes = list(movie_node_names)
# mediated_nodes = list(mediated_node_names)
# ground_truth_node = { ... }  # dictionary mapping mediated nodes to original nodes
# ground_truth_df = create_ground_truth_df(original_nodes, mediated_nodes, ground_truth_node)
# print(ground_truth_df)
```

```python
def max_pooling(matrices):
    # Convert matrices to numpy arrays and perform max pooling
    matrices_array = np.array([matrix.fillna(0).values for matrix in matrices])
    return pd.DataFrame(np.maximum.reduce(matrices_array), index=matrices[0].index, columns=matrices[0].columns)


def min_pooling(matrices):
    return np.minimum.reduce(matrices)

def mean_pooling(matrices):
    return np.mean(matrices, axis=0)

def weighted_pooling(matrices):
    std_devs = []
    for df in matrices:
      std_dev = df.std().mean()
      std_devs.append(std_dev)
    total_std_dev = sum(std_devs)
    weights = [std_dev / total_std_dev for std_dev in std_devs]
    matrices_array = np.array(matrices)
    assert len(matrices) == len(weights)
    weighted_average = np.average(matrices_array, axis=0, weights=weights)
    return weights, weighted_average
```

```python
# Main pooling function that returns a DataFrame
def pool_matrices_to_dataframe(similarity_matrices, pooling_type, source_columns, mediated_columns, weights=None):
    if pooling_type == "max":
        pooled_matrix = max_pooling(similarity_matrices)
    elif pooling_type == "min":
        pooled_matrix = min_pooling(similarity_matrices)
    elif pooling_type == "mean" or pooling_type == "average":
        pooled_matrix = mean_pooling(similarity_matrices)
    elif pooling_type == "weighted":
        weights, pooled_matrix = weighted_pooling(similarity_matrices)
    else:
        raise ValueError("Invalid pooling type. Choose 'max', 'mean', 'average', or 'weighted'.")

    return weights,pd.DataFrame(pooled_matrix, index=source_columns, columns=mediated_columns)
```

```python
def evaluate_node_similarity_results(similarity_results, ground_truth, threshold=0):
    tp = 0
    fp = 0
    fn = 0

    for path1 in similarity_results.index:
        best_match = similarity_results.loc[path1].idxmax()
        best_similarity = similarity_results.loc[path1].max()

        if best_similarity >= threshold:
            if ground_truth.get(path1) == best_match:
                tp += 1
            else:
                fp += 1
        elif ground_truth.get(path1) == best_match:
            fn += 1

    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    return {
        'TP': tp,
        'FP': fp,
        'FN': fn,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1_score
    }
```
